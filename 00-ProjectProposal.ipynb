## Research Question

How does the distribution of birth months of the 2022 and 2026 US Winter Olympic athletes differ from the general American population, based off of the 2000 census?

## Background and Prior Work

Every four years, many athletes of snow and ice sports such as figure skating, skiing, and curling participate in a major international multi-sport event known as the Winter Olympics. The competitors who reach this stage are considered elite athletes, representing the highest level of performance in their respective sports. However, the development of elite athletes may be influenced by external factors before they reach the Olympic stage. One such factor is the relative age effect (RAE), which suggests that athletes born earlier in the calendar year may have performance advantages in age-restricted sports leagues <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Because these athletes are often slightly older than their peers, they may be more physically and emotionally mature and thus more likely selected during early selection processes for those sports. Overtime, this advantage can contribute to them being more advanced athletes.

The first study to show the relative age effect was Barnsley, Thompson, and Barnsley in 1985. They looked at top junior and pro hockey leagues in Canada. They found that way more players were born in January, February, and March than in October, November, and December.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Many later studies looked for the same pattern in other sports. In 2009, Cobley and his team did a review that combined 38 studies across 14 sports and 16 countries. They found that for every 2 elite athletes born in the last three months of the year, there were about 3 born in the first three months. The effect was small but showed up almost everywhere.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) A 2020 study by Joyner and his team checked Olympic medal winners from 2000 to 2018 and found the RAE shows up even at the top level. Athletes born in spring were more likely to win gold than athletes born in winter (Q4).<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Raschner and his team in 2012 looked at the first Winter Youth Olympic Games. They found a strong RAE in skiing, snowboarding, and Nordic combined, but almost none in figure skating, which suggests the effect depends a lot on the sport.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

This is because in all of the prior works that we researched, the relative age effect applied for most prior Winter Olympic athletes. As such, we expect for this to be applied to recent Winter Olympic athletes as well.
