## Research Question

How does the distribution of birth months of the 2022 and 2026 US Winter Olympic athletes differ from the general American population, based off of the 2000 census?

## Background and Prior Work

Every four years, many athletes of snow and ice sports such as figure skating, skiing, and curling participate in a major international multi-sport event known as the Winter Olympics. The competitors who reach this stage are considered elite athletes, representing the highest level of performance in their respective sports. However, the development of elite athletes may be influenced by external factors before they reach the Olympic stage. One such factor is the relative age effect (RAE), which suggests that athletes born earlier in the calendar year may have performance advantages in age-restricted sports leagues <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Because these athletes are often slightly older than their peers, they may be more physically and emotionally mature and thus more likely selected during early selection processes for those sports. Overtime, this advantage can contribute to them being more advanced athletes.

The first study to show the relative age effect was Barnsley, Thompson, and Barnsley in 1985. The study took birth records of all the 1982/83 players from the Western Hockey League, Ontario Hockey League, and National Hockey Leagues, which are top junior and pro hockey leagues in Canada. By tabulating each of the birthdates by month, they found that players were twice as likely to be born in January, February, and March than in October, November, and December.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

Many later studies looked for the same pattern in other sports. In 2009, Cobley and his team did a review that combined 38 studies across 14 sports and 16 countries. They found that for every 2 elite athletes born in the last three months of the year, there were about 3 born in the first three months. The effect was small but showed up almost everywhere.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) A 2020 study by Joyner and his team checked Olympic medal winners from 2000 to 2018 and found the RAE shows up even at the top level. Athletes born in spring were more likely to win gold than athletes born in winter (Q4).<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Raschner and his team in 2012 looked at the first Winter Youth Olympic Games. They found a strong RAE in skiing, snowboarding, and Nordic combined, but almost none in figure skating, which suggests the effect depends a lot on the sport.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

Since these studies had populations of athletes that competed more than 8 years ago, we suspect that some of the results may have changed.

### Dataset #1: Team USA Winter Olympic Athletes, Birthdays and Event Results (2022 + 2026)

We made this dataset ourselves by scraping olympedia.org, which is the most complete free database of Olympic athletes online. We ran our own python scripts (`scraping.py` and `scraping_results.py`) on May 5, 2026 to pull the Team USA pages for the 2022 Beijing Winter Olympics and the 2026 Milano-Cortina Winter Olympics. For each Games we got two files: a "birthdays" file with one row per athlete and a "results" file with one row per event each athlete competed in. So in total there are four CSV files in `data/00-raw/`.

The birthdays files have the columns `id`, `name`, `born`, `birthplace`, and `url`. The most important one is `born`, which is a string that looks like "27 March 1992", and we parse it into a real date with pandas so we can pull out the birth month, which is the variable that our hypothesis is about. There are 223 athletes in 2022 and 230 in 2026 (so 453 total rows). Some athletes went to both Games, so the unique athlete count is actually smaller, about 361.

The results files have the columns `sport`, `event`, `athlete_id`, `athlete`, `position`, `medal`, and `is_team_entry`. The `position` column is normally a number like "1" or "14", but some of them are strings like "AC" (also competed) or "DNF" (did not finish), so when we convert it to a number those become NaN. The `medal` column is "Gold", "Silver", "Bronze", or empty. The `is_team_entry` column is "yes" for the team-level row in hockey or relays where no individual athlete is linked, and "no" for normal athlete rows. We drop the "yes" rows when we summarize per athlete because they would double count the medal. There are 225 result rows for 2022 and 233 for 2026.

After we load and clean the data we build a few new columns to actually do our analysis with. `birth_month` (1-12), `birth_quarter` (1-4), and `birth_year` come from parsing the `born` string. `age_at_games` is how old the athlete was at the opening ceremony of their Games (Feb 4 2022 for Beijing, Feb 6 2026 for Milano-Cortina). And then from the results file we build per athlete per Games summaries: how many events they entered, how many medals they won, whether they won any gold/silver/bronze, what their best finishing position was, and what sport(s) they did.

**Limitations of the dataset:**

- These are only the athletes who actually made the Olympic team. So we are looking at the survivors of a long selection process and we can't see all the kids who were filtered out earlier. The relative age effect might be much stronger in the youth pipeline than what shows up here.
- The sample is pretty small. With only around 450 athlete-Games rows and 12 months, doing a chi-square test by month gets really noisy, so we mostly look at quarters (4 buckets) instead.
- Olympedia is a community-run wiki so the data could have small inconsistencies, but when we checked the birthdays they all parsed fine with zero missing.
- Different winter sports use different selection year cutoffs. Most US winter sports use Jan 1 but a few don't. We are pooling all sports together for the main analysis, so this probably makes any relative age effect we find look smaller than it really is.
- Athletes who went to both Games are counted twice in the 453-row dataset. That's the right thing if you're counting athlete-Games appearances, but it slightly inflates the sample for our analysis. We flag this in case we want to look at unique athletes only.


```python
import pandas as pd
import os

# load all four CSVs
b22 = pd.read_csv("data/00-raw/olympedia_birthdays_2022.csv")
b26 = pd.read_csv("data/00-raw/olympedia_birthdays_2026.csv")
r22 = pd.read_csv("data/00-raw/olympedia_results_2022.csv")
r26 = pd.read_csv("data/00-raw/olympedia_results_2026.csv")

# tag each one with the year so we know which Games the row came from
b22["games_year"] = 2022
b26["games_year"] = 2026
r22["games_year"] = 2022
r26["games_year"] = 2026

birthdays = pd.concat([b22, b26], ignore_index=True)
results = pd.concat([r22, r26], ignore_index=True)

print("birthdays rows:", len(birthdays))
print("results rows:", len(results))

# the 'born' column is a string like "27 March 1992", parse it to a real date
birthdays["birth_date"] = pd.to_datetime(birthdays["born"], dayfirst=True, errors="coerce")
birthdays["birth_year"] = birthdays["birth_date"].dt.year
birthdays["birth_month"] = birthdays["birth_date"].dt.month

# birth quarter: 1=Jan-Mar, 2=Apr-Jun, 3=Jul-Sep, 4=Oct-Dec
quarters = []
for m in birthdays["birth_month"]:
    if pd.isna(m):
        quarters.append(None)
    elif m <= 3:
        quarters.append(1)
    elif m <= 6:
        quarters.append(2)
    elif m <= 9:
        quarters.append(3)
    else:
        quarters.append(4)
birthdays["birth_quarter"] = quarters

# how old were they at the opening ceremony of their Games
opening_2022 = pd.Timestamp("2022-02-04")
opening_2026 = pd.Timestamp("2026-02-06")
ages = []
for i in range(len(birthdays)):
    yr = birthdays.iloc[i]["games_year"]
    bd = birthdays.iloc[i]["birth_date"]
    if pd.isna(bd):
        ages.append(None)
    elif yr == 2022:
        ages.append(round((opening_2022 - bd).days / 365.25, 2))
    else:
        ages.append(round((opening_2026 - bd).days / 365.25, 2))
birthdays["age_at_games"] = ages

# id has to be a string to merge with athlete_id later
birthdays["id"] = birthdays["id"].astype(str)

birthdays[["id","name","games_year","born","birth_date","birth_month","birth_quarter","age_at_games"]].head()
```

```python
# check for missing values
print("missing values in birthdays:")
print(birthdays.isnull().sum())
print()

# check for any rows where the date didn't parse
bad_dates = birthdays[birthdays["birth_date"].isnull()]
print("rows where born string didn't parse:", len(bad_dates))
if len(bad_dates) > 0:
    print(bad_dates[["id","name","games_year","born"]])

# check for duplicates - shouldn't have the same athlete listed twice for the same Games
dupes = birthdays.duplicated(subset=["id","games_year"]).sum()
print("\nduplicate (id, year) rows:", dupes)

# check how many athletes were at BOTH Games (expected, not a problem)
both = 0
for athlete_id in birthdays["id"].unique():
    these = birthdays[birthdays["id"] == athlete_id]
    if len(these["games_year"].unique()) == 2:
        both += 1
print("athletes who went to BOTH 2022 and 2026:", both)

# look at the age range, anything younger than 14 or older than 55 is sus
print("\nage range:")
print("min:", birthdays["age_at_games"].min())
print("max:", birthdays["age_at_games"].max())
weird_ages = birthdays[(birthdays["age_at_games"] < 14) | (birthdays["age_at_games"] > 55)]
print("athletes with weird ages:", len(weird_ages))
if len(weird_ages) > 0:
    print(weird_ages[["name","games_year","age_at_games"]])

# drop any rows where we couldn't get a birth month, we need that for the analysis
# (turns out there are zero, but we do it anyway just in case)
birthdays_clean = birthdays.dropna(subset=["birth_month"]).copy()
print("\nbirthdays_clean rows:", len(birthdays_clean))
```

```python
# clean up the results dataframe
results["medal"] = results["medal"].fillna("").astype(str).str.strip()
# try to convert position to a number — "AC", "DNF" etc become NaN
results["position_num"] = pd.to_numeric(results["position"], errors="coerce")

# drop team-level rows (where it's just "United States" with no individual athlete)
indiv = results[results["is_team_entry"] == "no"].copy()
indiv["athlete_id"] = indiv["athlete_id"].astype(int).astype(str)
print("individual result rows:", len(indiv))

# build a summary: one row per (athlete, Games) with stats about how they did
summary_rows = []
for (aid, yr), group in indiv.groupby(["athlete_id","games_year"]):
    medals_list = group["medal"].tolist()
    gold = "Gold" in medals_list
    silver = "Silver" in medals_list
    bronze = "Bronze" in medals_list
    medal_count = 0
    for m in medals_list:
        if m in ["Gold","Silver","Bronze"]:
            medal_count += 1
    sports_list = sorted(set(group["sport"].tolist()))
    summary_rows.append({
        "athlete_id": aid,
        "games_year": yr,
        "events_entered": group["event"].nunique(),
        "medal_count": medal_count,
        "any_gold": gold,
        "any_silver": silver,
        "any_bronze": bronze,
        "any_medal": gold or silver or bronze,
        "best_position": group["position_num"].min(),
        "sports": ", ".join(sports_list),
    })
summary = pd.DataFrame(summary_rows)
print("summary rows:", len(summary))

# merge the birthdays with the summary so every athlete-Games has their stats
master = birthdays_clean.merge(
    summary,
    left_on=["id","games_year"],
    right_on=["athlete_id","games_year"],
    how="left"
)
master = master.drop(columns=["athlete_id"])

# some athletes (mostly the hockey team) have no individual results, only team rows
# we keep them anyway because we still want their birth month for the analysis
master["has_individual_results"] = master["events_entered"].notna()
master["events_entered"] = master["events_entered"].fillna(0).astype(int)
master["medal_count"] = master["medal_count"].fillna(0).astype(int)
master["any_gold"] = master["any_gold"].fillna(False)
master["any_silver"] = master["any_silver"].fillna(False)
master["any_bronze"] = master["any_bronze"].fillna(False)
master["any_medal"] = master["any_medal"].fillna(False)

print("master rows:", len(master))
print("athletes with no individual results:", (~master["has_individual_results"]).sum())
print("\nmedal-winning appearances per Games:")
print(master.groupby("games_year")["any_medal"].sum())

# make the output folders if they don't exist
os.makedirs("data/01-interim", exist_ok=True)
os.makedirs("data/02-processed", exist_ok=True)

# save intermediate steps
birthdays_clean.to_csv("data/01-interim/athletes_birthdays_combined.csv", index=False)
indiv.to_csv("data/01-interim/athlete_results_individual.csv", index=False)
summary.to_csv("data/01-interim/athlete_games_summary.csv", index=False)

# save the final analysis-ready dataset
cols = ["id","name","games_year","birth_date","birth_year","birth_month","birth_quarter",
        "age_at_games","birthplace","sports","events_entered","medal_count","any_medal",
        "any_gold","any_silver","any_bronze","best_position","has_individual_results"]
master_out = master[cols].sort_values(["games_year","name"]).reset_index(drop=True)
master_out.to_csv("data/02-processed/team_usa_winter_olympians.csv", index=False)
print("\nfiles saved!")

# quick summary stats for the hypothesis
print("\nbirth quarter distribution:")
qcounts = master_out["birth_quarter"].value_counts().sort_index()
qpct = (master_out["birth_quarter"].value_counts(normalize=True).sort_index() * 100).round(1)
print(pd.DataFrame({"count": qcounts, "percent": qpct.astype(str) + "%"}))

print("\nbirth month distribution:")
mcounts = master_out["birth_month"].value_counts().sort_index()
print(mcounts)

print("\ncrosstab of Games year x birth quarter:")
print(pd.crosstab(master_out["games_year"], master_out["birth_quarter"], margins=True))
```
