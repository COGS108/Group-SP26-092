## Abstract
Previous studies have noted an external factor known as the relative age effect (RAE) that suggests athletes born earlier in the calendar year may have performance advantages in age-restricted sports leagues. Since most prior research was done on athletes from over ten years ago, we want to study whether the RAE still holds true by analyzing birth distributions between the 2022 and 2026 Winter Olympic athletes and the general American population. Because these athletes are often slightly older than their peers, studies suggest that they may be more physically and emotionally mature and thus more likely to be selected during early selection processes for those sports. As such, we hypothesize that there will be a higher proportion of US Winter Olympic athletes from the 2022 and 2026 Winter Olympics born in the earlier calendar months (January to March) compared to the average US population. In order to complete this research, we needed to use two datasets: one for the birth months of the Winter Olympic athletes obtained by web-scraping olympedia.org and one for the birth months of the general population from a FiveThirtyEight dataset as a CSV file. By using a correlational design, we compared the distribution of first and second datasets to see whether there were more athletes born in the earlier birth months than the average population. From our chi-square test, we concluded that our hypothesis was not statistically significant. Therefore, our research shows that there is no clear evidence that there are more 2022 and 2026 Winter Olympic athletes born in the earlier months compared to the general population.

## Background and Prior Work

Every four years, many athletes of snow and ice sports such as figure skating, skiing, and curling participate in a major international multi-sport event known as the Winter Olympics. The competitors who reach this stage are considered elite athletes, representing the highest level of performance in their respective sports. However, the development of elite athletes may be influenced by external factors before they reach the Olympic stage. One such factor is the relative age effect (RAE), which suggests that athletes born earlier in the calendar year may have performance advantages in age-restricted sports leagues <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Because these athletes are often slightly older than their peers, they may be more physically and emotionally mature and thus more likely selected during early selection processes for those sports. Overtime, this advantage can contribute to them being more advanced athletes.

The first study to show the relative age effect was Barnsley, Thompson, and Barnsley in 1985. The study took birth records of all the 1982/83 players from the Western Hockey League, Ontario Hockey League, and National Hockey Leagues, which are top junior and pro hockey leagues in Canada. By tabulating each of the birthdates by month, they found that players were twice as likely to be born in January, February, and March than in October, November, and December.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) However, this study has limitations because it was conducted over 40 years ago, and the population only included hockey players.

Many later studies looked for the same pattern in other sports. In 2009, Cobley and his team did a review that combined 38 studies conducted between 1984 and 2007 across 14 sports and 16 countries. This study measured Relative Age Effects (RAEs) by comparing athletes’ birthdates within annual age-grouped systems. The researchers evaluated whether athletes born earlier in the year were overrepresented in comparison to athletes born later. They found that for every 2 elite athletes born in the last three months of the year, there were about 3 born in the first three months. The effect was small but showed up almost everywhere. One limitation of this study is that most of the samples focused heavily on male athletes and Western sport systems. In addition, because this study relied on past published research, their differences in methods and participant selection could have affected consistency.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

A 2020 study by Joyner and his team explored whether Relative Age Effects (RAEs) continue beyond youth sports and remain present among elite Olympic athletes. The researchers examined over 44,000 Olympic athletes born between 1964 and 1996 using data collected from the International Olympic Committee database. The population included both male and female athletes competing in summer and winter Olympic sports, as well as athletes participating in team versus individual sports and ball versus nonball sports. They found the RAE shows up even at the top level. Athletes born in spring were more likely to win gold than athletes born in winter (Q4). One limitation of this study is that it grouped athletes broadly by Olympic category rather than by each individual sport. Because different sports have different demands and processes, this grouping may have masked how strong the RAE was present in certain sports. Additionally, the study did not distinguish between prepubescent and postpubescent athletes, particularly among female competitors, which may affect interpretations of maturity-related advantages.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

Raschner and his team in 2012 looked at the first Winter Youth Olympic Games. The researchers examined 1,021 athletes between the ages of 14 and 19, including 557 males and 464 females, who competed across 15 winter sports. Because the Youth Olympic Games grouped athletes within two-year age ranges, competitors could differ in age by nearly two full years, creating the potential for developmental advantages among relatively older athletes. The study measured RAE by analyzing athletes’ birthdates and comparing the observed birth distribution to an expected equal distribution using chi-square statistical tests. The researchers also collected anthropometric data, including height and weight, and used ANOVA and Kruskal-Wallis analyses to determine whether physical differences existed between relatively older and younger athletes. The study found a strong RAE in skiing, snowboarding, and Nordic combined, but almost none in figure skating, which suggests the effect depends a lot on the sport. One limitation of the study is that it focused only on athletes competing at a single Youth Olympic Games event, which limits the ability to generalize findings across different competitions, countries, or age-group systems. Another limitation is that the study primarily examined birthdate distributions and physical measurements without directly measuring psychological, social, or coaching-related factors that may also contribute to athlete selection and development.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

Since these studies had populations of athletes that competed more than 8 years ago, we suspect that some of the results may have changed.

The results files have the columns `sport`, `event`, `athlete_id`, `athlete`, `position`, `medal`, and `is_team_entry`. The `position` column is normally a number like "1" or "14", but some of them are strings like "AC" (also competed) or "DNF" (did not finish), so when we convert it to a number those become NaN. The `medal` column is "Gold", "Silver", "Bronze", or empty. The `is_team_entry` column is "yes" for the team-level row in hockey or relays where no individual athlete is linked, and "no" for normal athlete rows. We drop the "yes" rows when we summarize per athlete because they would double count the medal. There are 225 result rows for 2022 and 233 for 2026.

After we load and clean the data we build a few new columns to actually do our analysis with. `birth_month` (1-12), `birth_quarter` (1-4), and `birth_year` come from parsing the `born` string. `age_at_games` is how old the athlete was at the opening ceremony of their Games (Feb 4 2022 for Beijing, Feb 6 2026 for Milano-Cortina). And then from the results file we build per athlete per Games summaries: how many events they entered, how many medals they won, whether they won any gold/silver/bronze, what their best finishing position was, and what sport(s) they did.

**Limitations of the dataset:**

- These are only the athletes who actually made the Olympic team. So we are looking at the survivors of a long selection process and we can't see all the kids who were filtered out earlier. The relative age effect might be much stronger in the youth pipeline than what shows up here.
- The sample is pretty small. With only around 450 athlete-Games rows and 12 months, doing a chi-square test by month gets really noisy, so we mostly look at quarters (4 buckets) instead.
- Olympedia is a community-run wiki so the data could have small inconsistencies, but when we checked the birthdays they all parsed fine with zero missing.
- Different winter sports use different selection year cutoffs. Most US winter sports use Jan 1 but a few don't. We are pooling all sports together for the main analysis, so this probably makes any relative age effect we find look smaller than it really is.
- Athletes who went to both Games are counted twice in the 453-row dataset. That's the right thing if you're counting athlete-Games appearances, but it slightly inflates the sample for our analysis. We flag this in case we want to look at unique athletes only.


```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# load all four CSVs
b22 = pd.read_csv("data/00-raw/olympedia_birthdays_2022.csv")
b26 = pd.read_csv("data/00-raw/olympedia_birthdays_2026.csv")
r22 = pd.read_csv("data/00-raw/olympedia_results_2022.csv")
r26 = pd.read_csv("data/00-raw/olympedia_results_2026.csv")

# tag each one with the year so we know which Games the row came from
b22["games_year"] = 2022
b26["games_year"] = 2026
r22["games_year"] = 2022
r26["games_year"] = 2026

birthdays = pd.concat([b22, b26], ignore_index=True)
results = pd.concat([r22, r26], ignore_index=True)

print("birthdays rows:", len(birthdays))
print("results rows:", len(results))

# the 'born' column is a string like "27 March 1992", parse it to a real date
birthdays["birth_date"] = pd.to_datetime(birthdays["born"], dayfirst=True, errors="coerce")
birthdays["birth_year"] = birthdays["birth_date"].dt.year
birthdays["birth_month"] = birthdays["birth_date"].dt.month

# birth quarter: 1=Jan-Mar, 2=Apr-Jun, 3=Jul-Sep, 4=Oct-Dec
quarters = []
for m in birthdays["birth_month"]:
    if pd.isna(m):
        quarters.append(None)
    elif m <= 3:
        quarters.append(1)
    elif m <= 6:
        quarters.append(2)
    elif m <= 9:
        quarters.append(3)
    else:
        quarters.append(4)
birthdays["birth_quarter"] = quarters

# how old were they at the opening ceremony of their Games
opening_2022 = pd.Timestamp("2022-02-04")
opening_2026 = pd.Timestamp("2026-02-06")
ages = []
for i in range(len(birthdays)):
    yr = birthdays.iloc[i]["games_year"]
    bd = birthdays.iloc[i]["birth_date"]
    if pd.isna(bd):
        ages.append(None)
    elif yr == 2022:
        ages.append(round((opening_2022 - bd).days / 365.25, 2))
    else:
        ages.append(round((opening_2026 - bd).days / 365.25, 2))
birthdays["age_at_games"] = ages

# id has to be a string to merge with athlete_id later
birthdays["id"] = birthdays["id"].astype(str)

birthdays[["id","name","games_year","born","birth_date","birth_month","birth_quarter","age_at_games"]].head()
```

```python
# check for missing values
print("missing values in birthdays:")
print(birthdays.isnull().sum())
print()

# check for any rows where the date didn't parse
bad_dates = birthdays[birthdays["birth_date"].isnull()]
print("rows where born string didn't parse:", len(bad_dates))
if len(bad_dates) > 0:
    print(bad_dates[["id","name","games_year","born"]])

# check for duplicates - shouldn't have the same athlete listed twice for the same Games
dupes = birthdays.duplicated(subset=["id","games_year"]).sum()
print("\nduplicate (id, year) rows:", dupes)

# check how many athletes were at BOTH Games (expected, not a problem)
both = 0
for athlete_id in birthdays["id"].unique():
    these = birthdays[birthdays["id"] == athlete_id]
    if len(these["games_year"].unique()) == 2:
        both += 1
print("athletes who went to BOTH 2022 and 2026:", both)

# look at the age range, anything younger than 14 or older than 55 is sus
print("\nage range:")
print("min:", birthdays["age_at_games"].min())
print("max:", birthdays["age_at_games"].max())
weird_ages = birthdays[(birthdays["age_at_games"] < 14) | (birthdays["age_at_games"] > 55)]
print("athletes with weird ages:", len(weird_ages))
if len(weird_ages) > 0:
    print(weird_ages[["name","games_year","age_at_games"]])

# drop any rows where we couldn't get a birth month, we need that for the analysis
# (turns out there are zero, but we do it anyway just in case)
birthdays_clean = birthdays.dropna(subset=["birth_month"]).copy()
print("\nbirthdays_clean rows:", len(birthdays_clean))

# quick cleanliness check for the columns we need most
birthday_key_cols = ["id", "name", "games_year", "birth_date", "birth_month", "birth_quarter", "age_at_games"]
birthday_missing = birthdays_clean[birthday_key_cols].isnull().sum()
birthday_checks = pd.DataFrame({
    "check": ["rows", "missing key values", "duplicate athlete-Games", "bad dates", "weird ages"],
    "value": [len(birthdays_clean), birthday_missing.sum(), dupes, len(bad_dates), len(weird_ages)]
})
display(birthday_checks)

birthday_missing.plot(kind="bar", figsize=(8, 3), title="Missing values in athlete birthday key columns")
plt.ylabel("missing rows")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# clean up the results dataframe
results["medal"] = results["medal"].fillna("").astype(str).str.strip()
# try to convert position to a number — "AC", "DNF" etc become NaN
results["position_num"] = pd.to_numeric(results["position"], errors="coerce")

# drop team-level rows (where it's just "United States" with no individual athlete)
indiv = results[results["is_team_entry"] == "no"].copy()
indiv["athlete_id"] = indiv["athlete_id"].astype(int).astype(str)
print("individual result rows:", len(indiv))

# build a summary: one row per (athlete, Games) with stats about how they did
summary_rows = []
for (aid, yr), group in indiv.groupby(["athlete_id","games_year"]):
    medals_list = group["medal"].tolist()
    gold = "Gold" in medals_list
    silver = "Silver" in medals_list
    bronze = "Bronze" in medals_list
    medal_count = 0
    for m in medals_list:
        if m in ["Gold","Silver","Bronze"]:
            medal_count += 1
    sports_list = sorted(set(group["sport"].tolist()))
    summary_rows.append({
        "athlete_id": aid,
        "games_year": yr,
        "events_entered": group["event"].nunique(),
        "medal_count": medal_count,
        "any_gold": gold,
        "any_silver": silver,
        "any_bronze": bronze,
        "any_medal": gold or silver or bronze,
        "best_position": group["position_num"].min(),
        "sports": ", ".join(sports_list),
    })
summary = pd.DataFrame(summary_rows)
print("summary rows:", len(summary))

# merge the birthdays with the summary so every athlete-Games has their stats
master = birthdays_clean.merge(
    summary,
    left_on=["id","games_year"],
    right_on=["athlete_id","games_year"],
    how="left"
)
master = master.drop(columns=["athlete_id"])

# some athletes (mostly the hockey team) have no individual results, only team rows
# we keep them anyway because we still want their birth month for the analysis
master["has_individual_results"] = master["events_entered"].notna()
master["events_entered"] = master["events_entered"].fillna(0).astype(int)
master["medal_count"] = master["medal_count"].fillna(0).astype(int)
master["any_gold"]   = master["any_gold"].where(master["any_gold"].notna(), other=False).astype(bool)
master["any_silver"] = master["any_silver"].where(master["any_silver"].notna(), other=False).astype(bool)
master["any_bronze"] = master["any_bronze"].where(master["any_bronze"].notna(), other=False).astype(bool)
master["any_medal"]  = master["any_medal"].where(master["any_medal"].notna(), other=False).astype(bool)

print("master rows:", len(master))
print("athletes with no individual results:", (~master["has_individual_results"]).sum())
print("\nmedal-winning appearances per Games:")
print(master.groupby("games_year")["any_medal"].sum())

# make the output folders if they don't exist
os.makedirs("data/01-interim", exist_ok=True)
os.makedirs("data/02-processed", exist_ok=True)

# save intermediate steps
birthdays_clean.to_csv("data/01-interim/athletes_birthdays_combined.csv", index=False)
indiv.to_csv("data/01-interim/athlete_results_individual.csv", index=False)
summary.to_csv("data/01-interim/athlete_games_summary.csv", index=False)

# fill sport names missed by the result scraper for nested team/pair rows
# Olympedia stores team rosters as nested rows, so the original results scraper missed
# many bobsleigh, curling, figure skating pair/dance, and ice hockey athletes.
supplement_path = "data/01-interim/athlete_sport_supplement.csv"
if os.path.exists(supplement_path):
    sport_supplement = pd.read_csv(supplement_path)
    sport_supplement["id"] = sport_supplement["id"].astype(str)
    master["id"] = master["id"].astype(str)
    master = master.merge(
        sport_supplement[["id", "games_year", "sport_from_roster"]],
        on=["id", "games_year"],
        how="left"
    )
    missing_sport = master["sports"].isna() | (master["sports"].astype(str).str.strip() == "")
    master["sports"] = master["sports"].where(~missing_sport, master["sport_from_roster"])
    filled_sports = missing_sport.sum() - (master["sports"].isna() | (master["sports"].astype(str).str.strip() == "")).sum()
    master = master.drop(columns=["sport_from_roster"])
    print("filled missing sports from roster supplement:", int(filled_sports))
else:
    print("sport supplement not found; leaving blank sports as-is")

# save the final analysis-ready dataset
cols = ["id","name","games_year","birth_date","birth_year","birth_month","birth_quarter",
        "age_at_games","birthplace","sports","events_entered","medal_count","any_medal",
        "any_gold","any_silver","any_bronze","best_position","has_individual_results"]
master_out = master[cols].sort_values(["games_year","name"]).reset_index(drop=True)
master_out.to_csv("data/02-processed/team_usa_winter_olympians.csv", index=False)
print("\nfiles saved!")

# final checks for the athlete dataset
master_key_cols = ["id", "name", "games_year", "birth_date", "birth_month", "birth_quarter", "age_at_games"]
master_missing = master_out[master_key_cols].isnull().sum()
master_checks = pd.DataFrame({
    "check": ["rows", "missing key values", "duplicate athlete-Games", "missing birth month", "missing birth quarter"],
    "value": [
        len(master_out),
        master_missing.sum(),
        master_out.duplicated(subset=["id", "games_year"]).sum(),
        master_out["birth_month"].isnull().sum(),
        master_out["birth_quarter"].isnull().sum()
    ]
})
display(master_checks)

master_missing.plot(kind="bar", figsize=(8, 3), title="Missing values in final athlete key columns")
plt.ylabel("missing rows")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# average age and standard deviation
print("\nage summary:")
print("average age:", round(master_out["age_at_games"].mean(), 2))
print("age standard deviation:", round(master_out["age_at_games"].std(), 2))

print("\nage summary by Games year:")
age_summary = master_out.groupby("games_year")["age_at_games"].agg(["mean", "std"]).round(2)
age_summary = age_summary.rename(columns={"mean": "average_age", "std": "age_standard_deviation"})
print(age_summary)

# quick summary stats for the hypothesis
print("\nbirth quarter distribution:")
qcounts = master_out["birth_quarter"].value_counts().sort_index()
qpct = (master_out["birth_quarter"].value_counts(normalize=True).sort_index() * 100).round(1)
print(pd.DataFrame({"count": qcounts, "percent": qpct.astype(str) + "%"}))

print("\nbirth month distribution:")
mcounts = master_out["birth_month"].value_counts().sort_index()
print(mcounts)

print("\ncrosstab of Games year x birth quarter:")
print(pd.crosstab(master_out["games_year"], master_out["birth_quarter"], margins=True))
```

### Dataset #2: Number of US Births by Months From 1995-2005

The dataset for the birth months of the general US population comes from FiveThirtyEight, which used to be a national polling organization. We use the CDC/NCHS file for 1995-2003 and the SSA file for 2004-2005. These files contain `year`, `month`, `date_of_month`, and `births`, which is the number of people born on that day given the `year`, `month`, and `date_of_month`. 

In order to create our own dataset, we summed up the number of `births` from each month to create a monthly distribution and a birth quarter distribution for 1995 to 2005.

**Limitations**
- The dataset did not come from a census; it is a non-governmental source which may limit its reliability.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# load the US births files
cdc_births = pd.read_csv("data/00-raw/US_births_1994-2003_CDC_NCHS.csv")
ssa_births = pd.read_csv("data/00-raw/US_births_2000-2014_SSA.csv")

# use CDC for 1995-2003 and SSA for 2004-2005
cdc_births = cdc_births[(cdc_births["year"] >= 1995) & (cdc_births["year"] <= 2003)].copy()
ssa_births = ssa_births[(ssa_births["year"] >= 2004) & (ssa_births["year"] <= 2005)].copy()

cdc_births["source"] = "CDC_NCHS"
ssa_births["source"] = "SSA"

us_births = pd.concat([cdc_births, ssa_births], ignore_index=True)

print("CDC rows:", len(cdc_births))
print("SSA rows:", len(ssa_births))
print("raw rows:", len(us_births))
print("years:", us_births["year"].min(), "to", us_births["year"].max())
print("columns:", list(us_births.columns))

# make a real date column
us_births["birth_date"] = pd.to_datetime(
    us_births["year"].astype(str) + "-" +
    us_births["month"].astype(str) + "-" +
    us_births["date_of_month"].astype(str),
    errors="coerce"
)

# check for missing stuff
print("\nmissing values:")
print(us_births.isnull().sum())

bad_dates = us_births[us_births["birth_date"].isnull()]
print("\nbad dates:", len(bad_dates))

dupes = us_births.duplicated(subset=["birth_date"]).sum()
print("duplicate dates:", dupes)

# drop rows that would mess up the month counts
us_births_clean = us_births.dropna(subset=["birth_date", "births"]).copy()
us_births_clean = us_births_clean.drop_duplicates(subset=["birth_date"])

# add month names
us_births_clean["month_name"] = us_births_clean["birth_date"].dt.month_name()

# add birth quarter
quarters = []
for m in us_births_clean["month"]:
    if m <= 3:
        quarters.append(1)
    elif m <= 6:
        quarters.append(2)
    elif m <= 9:
        quarters.append(3)
    else:
        quarters.append(4)
us_births_clean["birth_quarter"] = quarters

print("\nclean rows:", len(us_births_clean))
print("date range:", us_births_clean["birth_date"].min(), "to", us_births_clean["birth_date"].max())

# check if every year has the right number of days
year_days = us_births_clean.groupby("year")["birth_date"].count()
expected_days = []
for yr in year_days.index:
    start = pd.Timestamp(str(yr) + "-01-01")
    end = pd.Timestamp(str(yr) + "-12-31")
    expected_days.append((end - start).days + 1)
year_check = pd.DataFrame({
    "year": year_days.index,
    "days_observed": year_days.values,
    "days_expected": expected_days
})
year_check["missing_days"] = year_check["days_expected"] - year_check["days_observed"]
display(year_check)

# show missing values visually
birth_key_cols = ["year", "month", "date_of_month", "births", "source", "birth_date", "birth_quarter"]
birth_missing = us_births_clean[birth_key_cols].isnull().sum()
birth_checks = pd.DataFrame({
    "check": ["rows", "missing key values", "duplicate dates", "bad dates", "missing days"],
    "value": [len(us_births_clean), birth_missing.sum(), dupes, len(bad_dates), year_check["missing_days"].sum()]
})
display(birth_checks)

birth_missing.plot(kind="bar", figsize=(8, 3), title="Missing values in US births key columns")
plt.ylabel("missing rows")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

year_check.plot(x="year", y="days_observed", kind="bar", legend=False, figsize=(8, 3), title="Days observed for each year")
plt.ylabel("days")
plt.tight_layout()
plt.show()

# make monthly birth counts
month_births = us_births_clean.groupby("month")["births"].sum()
month_days = us_births_clean.groupby("month")["birth_date"].count()
month_names = us_births_clean.groupby("month")["month_name"].first()
month_quarters = us_births_clean.groupby("month")["birth_quarter"].first()

us_births_monthly = pd.DataFrame({
    "month": month_births.index,
    "month_name": month_names.values,
    "birth_quarter": month_quarters.values,
    "days_observed": month_days.values,
    "births": month_births.values
})
us_births_monthly = us_births_monthly.sort_values("month").reset_index(drop=True)
us_births_monthly["percent_births"] = (us_births_monthly["births"] / us_births_monthly["births"].sum() * 100).round(2)
us_births_monthly["avg_daily_births"] = (us_births_monthly["births"] / us_births_monthly["days_observed"]).round(1)

# make quarter birth counts
quarter_births = us_births_clean.groupby("birth_quarter")["births"].sum()
quarter_days = us_births_clean.groupby("birth_quarter")["birth_date"].count()

us_births_quarterly = pd.DataFrame({
    "birth_quarter": quarter_births.index,
    "days_observed": quarter_days.values,
    "births": quarter_births.values
})
us_births_quarterly = us_births_quarterly.sort_values("birth_quarter").reset_index(drop=True)
us_births_quarterly["quarter_label"] = "Q" + us_births_quarterly["birth_quarter"].astype(str)
us_births_quarterly["percent_births"] = (us_births_quarterly["births"] / us_births_quarterly["births"].sum() * 100).round(2)
us_births_quarterly["avg_daily_births"] = (us_births_quarterly["births"] / us_births_quarterly["days_observed"]).round(1)
us_births_quarterly = us_births_quarterly[["birth_quarter", "quarter_label", "days_observed", "births", "percent_births", "avg_daily_births"]]

# save the cleaned files
os.makedirs("data/02-processed", exist_ok=True)
us_births_monthly.to_csv("data/02-processed/us_births_monthly_distribution.csv", index=False)
us_births_quarterly.to_csv("data/02-processed/us_births_quarterly_distribution.csv", index=False)

print("\nmonthly distribution:")
display(us_births_monthly)

print("\nquarter distribution:")
display(us_births_quarterly)

print("files saved!")
```

How we use both datasets: Dataset #1 (2022 and 2026 Winter Olympic Athletes Births Per Month) is the experimental group which will show if there is a relationship between birth month and being an Olympic athlete. This is done by comparing with Dataset #2 (US Birth Dates From 1995-2005), which is the control group that establishes the normal distribution of birthdays. We will compare whether there is a significant difference in the number of US athletes born in the earlier birth months (as in Dataset #1) in comparison to the average birth months by the general US population (as in Dataset #2).

```python
import math
import re
from pathlib import Path

import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
from IPython.display import Image, display

pd.set_option('display.max_columns', None)
pio.renderers.default = 'notebook_connected'

MONTH_LABELS = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
QUARTER_LABELS = ['Q1', 'Q2', 'Q3', 'Q4']

# This is the older Plotly color language we liked: teal for athletes, copper for U.S. births,
# purple for model/expected references, red for below-expectation deviations, and quiet gray for guides.
COLORS = {
    'athletes': '#2f6f73',
    'us': '#d88c46',
    'expected': '#8c6bb1',
    'positive': '#2f6f73',
    'negative': '#c75d5d',
    'gray': '#687076',
    'light_gray': '#d8d2c4',
    'games_2022': '#4f7cac',
    'games_2026': '#c44e52',
}
QUARTER_COLORS = ['#4f7cac', '#2f6f73', '#8c6bb1', '#c44e52']

# Static PNGs live in data so GitHub and reports have durable images even when Plotly JS is unavailable.
FIGURES_DIR = Path('data/03-figures')
FIGURES_DIR.mkdir(parents=True, exist_ok=True)
for old_png in FIGURES_DIR.glob('*.png'):
    old_png.unlink()
_FIGURE_COUNTER = 1

STATE_ABBREVIATIONS = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
    'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'District of Columbia': 'DC',
    'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL',
    'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA',
    'Maine': 'ME', 'Maryland': 'MD', 'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN',
    'Mississippi': 'MS', 'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV',
    'New Hampshire': 'NH', 'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY',
    'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK', 'Oregon': 'OR',
    'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC', 'South Dakota': 'SD',
    'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virginia': 'VA',
    'Washington': 'WA', 'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY',
}

athletes = pd.read_csv('data/02-processed/team_usa_winter_olympians.csv')
us_monthly = pd.read_csv('data/02-processed/us_births_monthly_distribution.csv')
us_quarterly = pd.read_csv('data/02-processed/us_births_quarterly_distribution.csv')

athletes['month_label'] = pd.Categorical(
    athletes['birth_month'].map(lambda month: MONTH_LABELS[int(month) - 1]),
    categories=MONTH_LABELS,
    ordered=True,
)
athletes['quarter_label'] = pd.Categorical(
    athletes['birth_quarter'].map(lambda quarter: f'Q{int(quarter)}'),
    categories=QUARTER_LABELS,
    ordered=True,
)
athletes['medal_status'] = athletes['any_medal'].map({True: 'Won medal', False: 'No medal'})
us_monthly['month_label'] = pd.Categorical(us_monthly['month_name'].str[:3], categories=MONTH_LABELS, ordered=True)
us_quarterly['quarter_label'] = pd.Categorical(us_quarterly['quarter_label'], categories=QUARTER_LABELS, ordered=True)

def athlete_month_distribution():
    counts = (
        athletes.groupby(['birth_month', 'month_label'], observed=True)
        .size()
        .rename('athlete_count')
        .reset_index()
        .sort_values('birth_month')
    )
    counts['athlete_percent'] = counts['athlete_count'] / counts['athlete_count'].sum() * 100
    out = counts.merge(us_monthly[['month', 'percent_births']], left_on='birth_month', right_on='month', how='left')
    out['expected_count'] = out['percent_births'] / 100 * len(athletes)
    out['count_difference'] = out['athlete_count'] - out['expected_count']
    out['expected_ratio'] = out['athlete_count'] / out['expected_count']
    return out

def athlete_quarter_distribution():
    counts = (
        athletes.groupby(['birth_quarter', 'quarter_label'], observed=True)
        .size()
        .rename('athlete_count')
        .reset_index()
        .sort_values('birth_quarter')
    )
    counts['athlete_percent'] = counts['athlete_count'] / counts['athlete_count'].sum() * 100
    out = counts.merge(us_quarterly[['birth_quarter', 'percent_births']], on='birth_quarter', how='left')
    out['expected_count'] = out['percent_births'] / 100 * len(athletes)
    out['count_difference'] = out['athlete_count'] - out['expected_count']
    out['expected_ratio'] = out['athlete_count'] / out['expected_count']
    return out

def sport_distribution():
    return (
        athletes['sports']
        .fillna('Missing sport')
        .value_counts()
        .rename_axis('sport')
        .reset_index(name='athlete_count')
        .sort_values('athlete_count', ascending=True)
    )

def medal_quarter_distribution():
    out = (
        athletes.groupby(['quarter_label', 'medal_status'], observed=True)
        .size()
        .rename('athlete_count')
        .reset_index()
    )
    out['percent_within_quarter'] = out['athlete_count'] / out.groupby('quarter_label', observed=True)['athlete_count'].transform('sum') * 100
    return out

def sport_quarter_distribution():
    out = (
        athletes.groupby(['sports', 'birth_quarter', 'quarter_label'], observed=True)
        .size()
        .rename('athlete_count')
        .reset_index()
        .sort_values(['sports', 'birth_quarter'])
    )
    out['percent_within_sport'] = out['athlete_count'] / out.groupby('sports')['athlete_count'].transform('sum') * 100
    return out

def parse_birth_state(birthplace):
    if not isinstance(birthplace, str) or '(USA)' not in birthplace:
        return None
    place = re.sub(r'\s*\(USA\)\s*$', '', birthplace).strip()
    parts = [part.strip() for part in place.split(',')]
    if len(parts) < 2:
        return None
    return STATE_ABBREVIATIONS.get(parts[-1])

def birth_state_distribution():
    out = athletes.copy()
    out['birth_state'] = out['birthplace'].map(parse_birth_state)
    return (
        out.dropna(subset=['birth_state'])
        .groupby('birth_state')
        .size()
        .rename('athlete_count')
        .reset_index()
        .sort_values('athlete_count', ascending=False)
    )

def linear_fit(x_values, y_values):
    # Lightweight OLS for a small diagnostic plot; avoids adding another dependency just for one line.
    frame = pd.DataFrame({'x': x_values, 'y': y_values}).dropna()
    if len(frame) < 2:
        return [], [], float('nan'), float('nan')
    x_mean = frame['x'].mean()
    y_mean = frame['y'].mean()
    denom = ((frame['x'] - x_mean) ** 2).sum()
    slope = 0.0 if math.isclose(denom, 0.0) else ((frame['x'] - x_mean) * (frame['y'] - y_mean)).sum() / denom
    intercept = y_mean - slope * x_mean
    x_line = [float(frame['x'].min()), float(frame['x'].max())]
    y_line = [intercept + slope * x for x in x_line]
    r_squared = float(frame['x'].corr(frame['y']) ** 2)
    return x_line, y_line, float(slope), r_squared

def apply_base_layout(fig, title, width=900, height=540):
    fig.update_layout(
        title={'text': title, 'x': 0.02, 'xanchor': 'left'},
        template='plotly_white',
        width=width,
        height=height,
        margin={'l': 70, 'r': 40, 't': 85, 'b': 65},
        font={'family': 'Arial, sans-serif', 'size': 14, 'color': '#233048'},
        title_font={'size': 22, 'color': '#233048'},
        legend_title_text='',
    )
    fig.update_xaxes(showgrid=True, gridcolor='rgba(35,48,72,0.12)', zeroline=False)
    fig.update_yaxes(showgrid=True, gridcolor='rgba(35,48,72,0.12)', zeroline=False)
    return fig

def save_and_show_plot(fig_spec, width=900, height=540):
    # The old plotly style returns plain figure dictionaries with title, description, data, and layout.
    global _FIGURE_COUNTER
    title = fig_spec['title']
    fig = go.Figure(data=fig_spec['data'], layout=fig_spec.get('layout', {}))
    fig = apply_base_layout(fig, title, width=fig_spec.get('width', width), height=fig_spec.get('height', height))
    filename = re.sub(r'[^a-z0-9]+', '_', title.lower()).strip('_')
    path = FIGURES_DIR / f'{_FIGURE_COUNTER:02d}_{filename}.png'
    fig.write_image(path, width=fig.layout.width, height=fig.layout.height, scale=2)
    _FIGURE_COUNTER += 1
    display(fig)
    display(Image(filename=str(path)))
    return path

def fig_month_percent(month_dist):
    return {
        'title': 'Birth Month Distribution: Team USA Winter Olympians vs U.S. Births',
        'description': 'Grouped bar chart comparing athlete birth-month percentages with U.S. birth percentages. This is the core monthly relative-age comparison.',
        'data': [
            {
                'type': 'bar',
                'name': 'Team USA athletes',
                'x': month_dist['month_label'].astype(str).tolist(),
                'y': month_dist['athlete_percent'].round(2).tolist(),
                'marker': {'color': COLORS['athletes']},
                'hovertemplate': '%{x}<br>Athletes: %{y:.2f}%<extra></extra>',
            },
            {
                'type': 'bar',
                'name': 'U.S. births baseline',
                'x': month_dist['month_label'].astype(str).tolist(),
                'y': month_dist['percent_births'].round(2).tolist(),
                'marker': {'color': COLORS['us']},
                'hovertemplate': '%{x}<br>U.S. births: %{y:.2f}%<extra></extra>',
            },
        ],
        'layout': {
            'barmode': 'group',
            'xaxis': {'title': 'Birth month'},
            'yaxis': {'title': 'Percent of group', 'ticksuffix': '%'},
        },
        'width': 980,
        'height': 540,
    }

def fig_quarter_percent(quarter_dist):
    return {
        'title': 'Birth Quarter Distribution',
        'description': 'Grouped bar chart comparing athlete birth-quarter percentages with the U.S. births baseline.',
        'data': [
            {
                'type': 'bar',
                'name': 'Team USA athletes',
                'x': quarter_dist['quarter_label'].astype(str).tolist(),
                'y': quarter_dist['athlete_percent'].round(2).tolist(),
                'marker': {'color': COLORS['athletes']},
                'hovertemplate': '%{x}<br>Athletes: %{y:.2f}%<extra></extra>',
            },
            {
                'type': 'bar',
                'name': 'U.S. births baseline',
                'x': quarter_dist['quarter_label'].astype(str).tolist(),
                'y': quarter_dist['percent_births'].round(2).tolist(),
                'marker': {'color': COLORS['us']},
                'hovertemplate': '%{x}<br>U.S. births: %{y:.2f}%<extra></extra>',
            },
        ],
        'layout': {
            'barmode': 'group',
            'xaxis': {'title': 'Birth quarter'},
            'yaxis': {'title': 'Percent of group', 'ticksuffix': '%', 'range': [0, 34]},
        },
        'width': 820,
        'height': 520,
    }

def fig_observed_minus_expected_month(month_dist):
    colors = [COLORS['positive'] if value >= 0 else COLORS['negative'] for value in month_dist['count_difference']]
    return {
        'title': 'Observed Athlete Counts Minus Expected Counts by Month',
        'description': 'Diverging bar chart of observed athlete counts minus counts expected from U.S. births. Positive bars indicate overrepresentation.',
        'data': [
            {
                'type': 'bar',
                'name': 'Observed - expected',
                'x': month_dist['month_label'].astype(str).tolist(),
                'y': month_dist['count_difference'].round(2).tolist(),
                'marker': {'color': colors},
                'hovertemplate': '%{x}<br>Observed - expected: %{y:.2f}<extra></extra>',
            }
        ],
        'layout': {
            'xaxis': {'title': 'Birth month'},
            'yaxis': {'title': 'Observed - expected athlete count', 'range': [-17, 15]},
            'shapes': [{'type': 'line', 'x0': -0.5, 'x1': 11.5, 'y0': 0, 'y1': 0, 'line': {'color': '#233048', 'width': 1}}],
            'showlegend': False,
        },
        'width': 920,
        'height': 520,
    }

def fig_sport_distribution(sports):
    return {
        'title': 'Athlete Count by Sport',
        'description': 'Horizontal bar chart showing athlete-Games rows by sport after the sport-label cleaning fix.',
        'data': [
            {
                'type': 'bar',
                'orientation': 'h',
                'name': 'Athletes',
                'x': sports['athlete_count'].tolist(),
                'y': sports['sport'].tolist(),
                'marker': {'color': COLORS['athletes']},
                'hovertemplate': '%{y}<br>Athletes: %{x}<extra></extra>',
            }
        ],
        'layout': {
            'xaxis': {'title': 'Athlete count', 'range': [0, sports['athlete_count'].max() + 14]},
            'yaxis': {'title': 'Sport'},
            'showlegend': False,
        },
        'width': 920,
        'height': 700,
    }

def fig_age_violin_by_quarter(athletes):
    data = []
    for quarter, color in zip(QUARTER_LABELS, QUARTER_COLORS):
        frame = athletes[athletes['quarter_label'].astype(str) == quarter]
        ages = frame['age_at_games'].round(2).tolist()
        quarters = [quarter] * len(ages)
        data.append({
            'type': 'violin',
            'name': quarter,
            'x': quarters,
            'y': ages,
            'box': {'visible': False},
            'meanline': {'visible': False},
            'points': False,
            'line': {'color': color, 'width': 1.5},
            'fillcolor': color,
            'opacity': 0.42,
            'hovertemplate': f'{quarter}<br>Age: %{{y:.2f}}<extra></extra>',
            'legendgroup': quarter,
        })
        # Overlay a narrow box plot at the same quarter category so the inner summary remains visible.
        data.append({
            'type': 'box',
            'name': quarter,
            'x': quarters,
            'y': ages,
            'width': 0.16,
            'marker': {'color': 'rgba(35,48,72,0.55)', 'size': 3},
            'line': {'color': '#233048', 'width': 1.2},
            'fillcolor': 'rgba(255,255,255,0.72)',
            'boxmean': True,
            'showlegend': False,
            'hovertemplate': f'{quarter}<br>Age: %{{y:.2f}}<extra></extra>',
            'legendgroup': quarter,
        })
    return {
        'title': 'Age Distribution Shape by Birth Quarter',
        'description': 'Violin plot showing the full age distribution inside each birth quarter, with an inner box summary restored for median, spread, and outliers.',
        'data': data,
        'layout': {'xaxis': {'title': 'Birth quarter', 'categoryorder': 'array', 'categoryarray': QUARTER_LABELS}, 'yaxis': {'title': 'Age at Games'}, 'violingap': 0.15, 'boxmode': 'overlay'},
        'width': 820,
        'height': 520,
    }

def fig_medal_by_quarter(medal_quarter):
    totals = (
        medal_quarter.groupby('quarter_label', observed=True)['athlete_count']
        .sum()
        .reindex(QUARTER_LABELS)
        .reset_index()
        .rename(columns={'athlete_count': 'total_count'})
    )
    medal_counts = (
        medal_quarter[medal_quarter['medal_status'] == 'Won medal']
        .set_index('quarter_label')['athlete_count']
        .reindex(QUARTER_LABELS)
        .fillna(0)
        .astype(int)
        .reset_index()
        .rename(columns={'athlete_count': 'medal_count'})
    )
    frame = totals.merge(medal_counts, on='quarter_label')
    frame['medal_share'] = frame['medal_count'] / frame['total_count'] * 100
    return {
        'title': 'Medal Outcomes by Birth Quarter',
        'description': 'Overlay bar chart: the full bar is all athlete-Games rows in a quarter, and the shaded inner bar is the subset with at least one medal.',
        'data': [
            {
                'type': 'bar',
                'name': 'All athlete-Games rows',
                'x': frame['quarter_label'].astype(str).tolist(),
                'y': frame['total_count'].tolist(),
                'marker': {'color': COLORS['light_gray']},
                'hovertemplate': '%{x}<br>Total rows: %{y}<extra></extra>',
            },
            {
                'type': 'bar',
                'name': 'Won medal',
                'x': frame['quarter_label'].astype(str).tolist(),
                'y': frame['medal_count'].tolist(),
                'customdata': frame['medal_share'].round(2).tolist(),
                'marker': {'color': COLORS['expected']},
                'hovertemplate': '%{x}<br>Won medal: %{y}<br>Share: %{customdata:.2f}%<extra></extra>',
            },
        ],
        'layout': {
            'barmode': 'overlay',
            'xaxis': {'title': 'Birth quarter'},
            'yaxis': {'title': 'Athlete-Games count'},
        },
        'width': 820,
        'height': 520,
    }

def fig_birth_state_choropleth(state_counts):
    return {
        'title': 'U.S. Birthplace Map for Team USA Winter Olympians',
        'description': 'Choropleth of U.S.-born athlete-Games rows by birth state. This gives geographic context because access to winter sports is not evenly distributed.',
        'data': [
            {
                'type': 'choropleth',
                'locationmode': 'USA-states',
                'locations': state_counts['birth_state'].tolist(),
                'z': state_counts['athlete_count'].tolist(),
                'colorscale': [[0, '#f4f8f7'], [1, COLORS['athletes']]],
                'colorbar': {'title': 'Athletes'},
                'hovertemplate': '%{location}<br>Athlete-Games: %{z}<extra></extra>',
            }
        ],
        'layout': {'geo': {'scope': 'usa', 'bgcolor': 'rgba(0,0,0,0)'}},
        'width': 900,
        'height': 580,
    }

month_dist = athlete_month_distribution()
quarter_dist = athlete_quarter_distribution()
sports = sport_distribution()
medal_quarter = medal_quarter_distribution()
sport_quarter = sport_quarter_distribution()
state_counts = birth_state_distribution()

summary = {
    'athlete_rows': f'{len(athletes):,}',
    'games_years': ', '.join(str(year) for year in sorted(athletes['games_year'].unique())),
    'q1': f"{quarter_dist.loc[quarter_dist['quarter_label'].astype(str) == 'Q1', 'athlete_percent'].iloc[0]:.1f}% athletes vs {quarter_dist.loc[quarter_dist['quarter_label'].astype(str) == 'Q1', 'percent_births'].iloc[0]:.1f}% U.S. births",
    'q2': f"{quarter_dist.loc[quarter_dist['quarter_label'].astype(str) == 'Q2', 'athlete_percent'].iloc[0]:.1f}% athletes vs {quarter_dist.loc[quarter_dist['quarter_label'].astype(str) == 'Q2', 'percent_births'].iloc[0]:.1f}% U.S. births",
    'q3': f"{quarter_dist.loc[quarter_dist['quarter_label'].astype(str) == 'Q3', 'athlete_percent'].iloc[0]:.1f}% athletes vs {quarter_dist.loc[quarter_dist['quarter_label'].astype(str) == 'Q3', 'percent_births'].iloc[0]:.1f}% U.S. births",
    'q4': f"{quarter_dist.loc[quarter_dist['quarter_label'].astype(str) == 'Q4', 'athlete_percent'].iloc[0]:.1f}% athletes vs {quarter_dist.loc[quarter_dist['quarter_label'].astype(str) == 'Q4', 'percent_births'].iloc[0]:.1f}% U.S. births",
    'max_month': f"{month_dist.loc[month_dist['athlete_count'].idxmax(), 'month_label']} ({int(month_dist['athlete_count'].max())} athletes)",
    'min_month': f"{month_dist.loc[month_dist['athlete_count'].idxmin(), 'month_label']} ({int(month_dist['athlete_count'].min())} athletes)",
}

graph_type_coverage = pd.DataFrame([
    ['Grouped bar chart', 'Birth Month and Birth Quarter Comparisons'],
    ['Horizontal bar chart', 'Athlete Count by Sport'],
    ['Diverging bar chart', 'Observed Minus Expected Counts by Month'],
    ['Violin plot', 'Age Distribution Shape by Birth Quarter'],
    ['Overlay bar chart', 'Medal Outcomes by Birth Quarter'],
    ['Choropleth map', 'U.S. Birth State Distribution'],
], columns=['Graph type', 'Where used in this EDA'])
```

### What distributions do your variables take?

```python
# Plot 1: Core athlete-vs-population comparison by birth month.
_ = save_and_show_plot(fig_month_percent(month_dist))
```

This figure is the starting point for the EDA because it compares the athletes' birth-month distribution to a reasonable baseline: the U.S. birth-month distribution. The goal is not just to count which months are common, but to ask whether Team USA Winter Olympians look different from what we would expect if they were born throughout the year in the same pattern as the general population. If there were no birth-timing pattern, the teal athlete bars and the orange U.S. birth bars would be fairly close across all months.

The graph shows that the athlete distribution does not perfectly match the U.S. baseline. January is noticeably lower among athletes, while several spring and summer months are higher. This is an important first finding because it suggests that the athlete sample has some unevenness by birth month. However, the pattern is not enough on its own to explain why the unevenness exists, so the next step is to simplify the month-level variation into broader birth quarters.

```python
# Plot 2: Quarter-level version of the core comparison for the relative-age hypothesis.
_ = save_and_show_plot(fig_quarter_percent(quarter_dist))
```

This figure groups the same birth timing information into quarters, which makes the overall pattern easier to interpret. Month-by-month graphs can be noisy because a single month may be unusually high or low, but quarters give a clearer view of whether athletes are concentrated in earlier or later parts of the year. This directly connects to our relative age effect question because many versions of the theory focus on advantages connected to where an athlete's birthday falls in a selection year.

The quarter-level result is surprising because it does not strongly support the original expectation that Q1 athletes would be overrepresented. Instead, Q1 and Q4 are lower than the U.S. birth baseline, while Q2 and Q3 are higher. This suggests that the Team USA Winter Olympian sample has a birth-timing pattern, but the pattern is not the simple early-year advantage we expected. From a data science perspective, this is useful because it prevents us from forcing the data to fit the hypothesis and instead pushes us to investigate other explanations.

```python
# Plot 3: Sport composition sanity check after filling the missed Olympedia sport labels.
_ = save_and_show_plot(fig_sport_distribution(sports))
```

This sport composition graph helps explain the structure of the dataset before we interpret the birth-timing results too broadly. Each row in the dataset represents an athlete-Games appearance, but those rows are not evenly spread across sports. Ice hockey contributes the largest number of rows, followed by freestyle skiing and snowboarding, while several other sports contribute much smaller groups.

This matters because the overall birth-month and birth-quarter patterns could partly reflect which sports dominate the roster. For example, if a large sport has a different birth-timing pattern than smaller sports, it could influence the overall distribution. This graph therefore acts as a check on the dataset's composition. It reminds us that our EDA is describing Team USA Winter Olympian rows as collected, not a perfectly balanced sample of every winter sport.

### Are there any outliers?

```python
outlier_checks = pd.DataFrame({
    'check': [
        'rows',
        'missing birth_month',
        'missing birth_quarter',
        'missing sports',
        'duplicate athlete-Games rows',
        'youngest age_at_games',
        'oldest age_at_games',
    ],
    'value': [
        len(athletes),
        athletes['birth_month'].isna().sum(),
        athletes['birth_quarter'].isna().sum(),
        (athletes['sports'].isna() | (athletes['sports'].astype(str).str.strip() == '')).sum(),
        athletes.duplicated(subset=['id', 'games_year']).sum(),
        athletes['age_at_games'].min(),
        athletes['age_at_games'].max(),
    ],
})
display(outlier_checks)

age_extremes = athletes.sort_values('age_at_games').head(5)[['name', 'games_year', 'age_at_games', 'sports', 'birth_month']]
age_extremes = pd.concat([
    age_extremes,
    athletes.sort_values('age_at_games', ascending=False).head(5)[['name', 'games_year', 'age_at_games', 'sports', 'birth_month']],
])
print('Youngest and oldest athlete-Games appearances:')
display(age_extremes)

print('Months with the largest observed-minus-expected differences:')
display(month_dist.sort_values('count_difference')[['month_label', 'athlete_count', 'expected_count', 'count_difference']].round(2))
```

```python
# Plot 4: Month-level over/underrepresentation relative to the U.S. births baseline.
_ = save_and_show_plot(fig_observed_minus_expected_month(month_dist))
```

This figure turns the baseline comparison into a more direct measure of overrepresentation and underrepresentation. Instead of comparing two percentages side by side, it subtracts the expected number of athletes for each month from the observed number of athletes. This is useful because it shows the size and direction of the difference in athlete counts, which is easier to interpret than small percentage gaps.

The strongest negative difference appears in January, meaning there are far fewer January-born athletes than expected from U.S. birth rates. June shows the largest positive difference, meaning there are many more June-born athletes than expected. More broadly, the middle months tend to be above expectation, while the beginning and end of the year are often below expectation. This supports the earlier conclusion that birth timing is uneven, but it also confirms that the observed pattern is not a straightforward Q1 advantage.

```python
# Plot 5: Violin plot to compare full age-distribution shape by birth quarter.
_ = save_and_show_plot(fig_age_violin_by_quarter(athletes))
```

This violin plot checks whether age might be interfering with the birth-quarter results. Since the dataset includes athletes from both the 2022 and 2026 Winter Games, athletes are not all the same age. If one birth quarter had a very different age distribution, then the birth-quarter pattern might partly reflect age differences rather than birth timing itself.

The distributions look broadly similar across the four quarters, with most athletes concentrated in a comparable age range. There are some older outliers, but they do not appear to be isolated in only one quarter. This makes the birth-quarter pattern more credible as a birth-timing pattern rather than just an age-composition artifact.

### Relationship between variables

### Section 2: Statistical Summaries and Analysis

This section quantifies the patterns from Section 1 using descriptive statistics, chi-square goodness-of-fit tests comparing athletes to the U.S. birth baseline, a proportion test for the core Q1 hypothesis, medal-rate confidence intervals, and sport-level breakdowns.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

try:
    from statsmodels.stats.proportion import proportions_ztest, proportion_confint
    HAS_STATSMODELS = True
except ImportError:
    HAS_STATSMODELS = False
    print("statsmodels not found; skipping z-test and Wilson CIs")

# reload processed data (safe to re-run)
athletes  = pd.read_csv("data/02-processed/team_usa_winter_olympians.csv")
us_monthly  = pd.read_csv("data/02-processed/us_births_monthly_distribution.csv")
us_quarterly = pd.read_csv("data/02-processed/us_births_quarterly_distribution.csv")

MONTH_LABELS   = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
QUARTER_LABELS = ["Q1","Q2","Q3","Q4"]

athletes["month_label"]   = athletes["birth_month"].map(lambda m: MONTH_LABELS[int(m)-1])
athletes["quarter_label"] = athletes["birth_quarter"].map(lambda q: f"Q{int(q)}")

print("=== Dataset Summary ===")
print(f"Total Athlete-Games Rows : {len(athletes)}")
print(f"Unique Athletes          : {athletes['id'].nunique()}")
print(f"Games Years              : {sorted(athletes['games_year'].unique())}")

print("\n=== Birth Month Descriptive Stats ===")
mc = athletes["birth_month"].value_counts().sort_index()
print(f"Mean Athletes/Month      : {mc.mean():.1f}")
print(f"Std Dev Per Month        : {mc.std():.1f}")
print(f"Most Athletes            : {MONTH_LABELS[mc.idxmax()-1]} ({mc.max()})")
print(f"Fewest Athletes          : {MONTH_LABELS[mc.idxmin()-1]} ({mc.min()})")

print("\n=== Age at Games Descriptive Stats ===")
age = athletes["age_at_games"]
print(f"Mean   : {age.mean():.2f}  |  Median : {age.median():.2f}")
print(f"Std    : {age.std():.2f}   |  IQR    : {age.quantile(.75)-age.quantile(.25):.2f}")
print(f"Range  : {age.min():.1f} – {age.max():.1f}")

print("\n=== Medal Summary ===")
print(f"Athlete-Games with any medal : {athletes['any_medal'].sum()}  ({athletes['any_medal'].mean()*100:.1f}%)")
print(athletes.groupby("games_year")["any_medal"]
      .agg(["sum","mean"])
      .rename(columns={"sum":"medals","mean":"rate"})
      .assign(rate=lambda d: d["rate"].mul(100).round(1)))
```

The summary confirms the data is clean and complete. With 453 athlete-Games rows and no missing birth months, every athlete can be placed in a birth quarter for the chi-square tests below.

```python
# Chi-square goodness-of-fit: birth MONTH vs U.S. baseline
obs_m = athletes["birth_month"].value_counts().sort_index().values.astype(float)
exp_props_m = us_monthly.sort_values("month")["percent_births"].values / 100.0
exp_m = exp_props_m * obs_m.sum()

chi2_m, p_m = stats.chisquare(f_obs=obs_m, f_exp=exp_m)
dof_m = len(obs_m) - 1

print(f"Chi-Square (monthly):  chi2 = {chi2_m:.4f},  p = {p_m:.4f},  df = {dof_m}")
sig = "significant" if p_m < 0.05 else "NOT significant"
print(f"  → {sig} at alpha=0.05")

chi2_contrib = (obs_m - exp_m)**2 / exp_m
month_df = pd.DataFrame({
    "month"            : MONTH_LABELS,
    "observed"         : obs_m.astype(int),
    "expected"         : exp_m.round(1),
    "obs_pct"          : (obs_m / obs_m.sum() * 100).round(2),
    "exp_pct"          : (exp_props_m * 100).round(2),
    "chi2_contribution": chi2_contrib.round(4),
})
display(month_df)

# Bar chart of chi2 contributions by month
fig, ax = plt.subplots(figsize=(9, 4))
colors = ["#2f6f73" if v >= 0 else "#c75d5d" for v in (obs_m - exp_m)]
ax.bar(MONTH_LABELS, chi2_contrib, color=colors, alpha=0.85)
ax.axhline(0, color="#233048", linewidth=0.8)
ax.set_xlabel("Birth month")
ax.set_ylabel("Chi-square contribution")
ax.set_title("Per-Month Chi-Square Contributions (athlete vs U.S. baseline)")
sns.despine(ax=ax)
plt.tight_layout()
plt.savefig("data/03-figures/chi2_by_month.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: data/03-figures/chi2_by_month.png")
```

The bar chart shows which months contribute most to any overall chi-square signal. Taller bars indicate months where athletes are furthest from the expected count. The direction (overrepresented vs underrepresented) is reflected in the color: teal = more athletes than expected, red = fewer. The highest (in the underrepresented direction) contribution to the overall chi-square signal is January, while the highest in the overrepresented direction is June.

```python
# Chi-square goodness-of-fit: birth QUARTER vs U.S. baseline
obs_q = athletes["birth_quarter"].value_counts().sort_index().values.astype(float)
exp_props_q = us_quarterly.sort_values("birth_quarter")["percent_births"].values / 100.0
exp_q = exp_props_q * obs_q.sum()

chi2_q, p_q = stats.chisquare(f_obs=obs_q, f_exp=exp_q)
dof_q = len(obs_q) - 1

print(f"Chi-Square (quarterly):  chi2 = {chi2_q:.4f},  p = {p_q:.4f},  df = {dof_q}")
sig = "significant" if p_q < 0.05 else "NOT significant"
print(f"  -> {sig} at alpha=0.05")

quarter_df = pd.DataFrame({
    "quarter"          : QUARTER_LABELS,
    "observed"         : obs_q.astype(int),
    "expected"         : exp_q.round(1),
    "obs_pct"          : (obs_q / obs_q.sum() * 100).round(2),
    "exp_pct"          : (exp_props_q * 100).round(2),
    "diff_pct_pts"     : ((obs_q/obs_q.sum() - exp_props_q)*100).round(2),
    "chi2_contribution": ((obs_q - exp_q)**2 / exp_q).round(4),
})
display(quarter_df)
```

The quarterly chi-square is the most direct test of the relative-age hypothesis, with the result indicating there is a significant difference between athlete and U.S. baseline birth quarter distribution. The `diff_pct_pts` column shows the direction: positive = athletes overrepresented, negative = athletes underrepresented relative to U.S. birth rates. We see that Q1 and Q4 show underrepresentation while Q2 and Q3 show overrepresentation, with Q2 and Q4 being the largest contributors to the chi-square statistic.

```python
# One-proportion z-test: is the Q1 fraction significantly above the U.S. baseline?
q1_n  = int((athletes["birth_quarter"] == 1).sum())
q4_n  = int((athletes["birth_quarter"] == 4).sum())
n_tot = len(athletes)
q1_us = float(us_quarterly.loc[us_quarterly["birth_quarter"]==1, "percent_births"].iloc[0]) / 100

print(f"Q1 athletes        : {q1_n}/{n_tot} ({q1_n/n_tot*100:.1f}%)")
print(f"U.S. Q1 baseline   : {q1_us*100:.1f}%")

if HAS_STATSMODELS:
    z, p_one = proportions_ztest(count=q1_n, nobs=n_tot, value=q1_us, alternative="larger")
    print(f"z = {z:.4f},  p (one-tailed) = {p_one:.4f}")
    sig = "Significant" if p_one < 0.05 else "Not significant"
    print(f"  -> {sig}: Q1 athletes are {'over' if p_one<0.05 else 'NOT over'}represented.")
else:
    # manual z-test fallback
    p_hat = q1_n / n_tot
    se = (q1_us * (1 - q1_us) / n_tot) ** 0.5
    z = (p_hat - q1_us) / se
    from scipy.stats import norm
    p_one = 1 - norm.cdf(z)
    print(f"z = {z:.4f},  p (one-tailed) = {p_one:.4f}")

print(f"\nQ1/Q4 ratio among athletes: {q1_n}/{q4_n} = {q1_n/q4_n:.2f}")
print("  (Barnsley 1985 hockey study found ratio ~2.0; value near 1.0 = no RAE)")
```

The one-tailed z-test directly tests the hypothesis: 'a higher proportion of US Winter Olympians are born in Q1 (Jan–Mar) than in the general US population.' The Q1/Q4 ratio is a secondary check drawn from the original Barnsley (1985) relative-age effect study. As no significant effect for Q1 was found, we will not conduct z-tests on individual months as it will increase the risk of getting false positives because of multiple comparisons.

```python
# Medal rate by birth quarter with 95% Wilson confidence intervals
medal_rows = []
for q in [1, 2, 3, 4]:
    sub = athletes[athletes["birth_quarter"] == q]
    n = len(sub)
    k = int(sub["any_medal"].sum())
    rate = k / n if n > 0 else 0
    if HAS_STATSMODELS:
        lo, hi = proportion_confint(k, n, alpha=0.05, method="wilson")
    else:
        # approximate Wilson CI
        z_ = 1.96
        denom = 1 + z_**2/n
        centre = (rate + z_**2/(2*n)) / denom
        margin = (z_ * (rate*(1-rate)/n + z_**2/(4*n**2))**0.5) / denom
        lo, hi = max(0, centre-margin), min(1, centre+margin)
    medal_rows.append({
        "quarter": f"Q{q}", "athletes": n, "medalists": k,
        "medal_rate_pct": round(rate*100,1),
        "ci_lower_pct":   round(lo*100,1),
        "ci_upper_pct":   round(hi*100,1),
    })
medal_rate_df = pd.DataFrame(medal_rows)
display(medal_rate_df)

fig, ax = plt.subplots(figsize=(7, 4))
x = range(4)
pal = ["#4f7cac","#2f6f73","#8c6bb1","#c44e52"]
ax.bar(x, medal_rate_df["medal_rate_pct"], color=pal, alpha=0.82, width=0.55)
ax.errorbar(x, medal_rate_df["medal_rate_pct"],
            yerr=[medal_rate_df["medal_rate_pct"]-medal_rate_df["ci_lower_pct"],
                  medal_rate_df["ci_upper_pct"]-medal_rate_df["medal_rate_pct"]],
            fmt="none", color="#233048", capsize=6, linewidth=1.5)
ax.axhline(athletes["any_medal"].mean()*100, color="#687076",
           linestyle="--", linewidth=1.2, label="Overall medal rate")
ax.set_xticks(list(x)); ax.set_xticklabels(QUARTER_LABELS)
ax.set_xlabel("Birth quarter"); ax.set_ylabel("Medal rate (%)")
ax.set_title("Medal Rate by Birth Quarter (95% Wilson CIs)")
ax.legend(); sns.despine(ax=ax)
plt.tight_layout()
plt.savefig("data/03-figures/medal_rate_ci.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: data/03-figures/medal_rate_ci.png")
```

Wide confidence intervals reflect the limited number of medalists per quarter. Overlapping CIs mean differences should be treated as exploratory only. This plot tests whether Q1 athletes not only make the Olympic team but also succeed once there. While Q1 athletes show a slightly higher medal rate than the overall medal rate, Q3 athletes show the highest medal rate. Nevertheless, the overlapping CIs between each quarter indicate considerable uncertainty regarding whether medal rate differs meaningfully by birth quarter.

```python
# Age distribution by birth quarter — violin + strip, and one-way ANOVA
fig, ax = plt.subplots(figsize=(8, 5))
QPAL = {"Q1":"#4f7cac","Q2":"#2f6f73","Q3":"#8c6bb1","Q4":"#c44e52"}
sns.violinplot(data=athletes, x="quarter_label", y="age_at_games",
               hue="quarter_label", palette=QPAL, order=QUARTER_LABELS,
               inner="box", linewidth=1.2, alpha=0.7, legend=False, ax=ax)
sns.stripplot(data=athletes, x="quarter_label", y="age_at_games",
              order=QUARTER_LABELS, color="#233048",
              alpha=0.25, size=3, jitter=True, ax=ax)
ax.set_xlabel("Birth quarter"); ax.set_ylabel("Age at Games opening ceremony")
ax.set_title("Age Distribution by Birth Quarter (violin + individual points)")
sns.despine(ax=ax); plt.tight_layout()
plt.savefig("data/03-figures/age_by_quarter_violin.png", dpi=150, bbox_inches="tight")
plt.show()

# ANOVA: does mean age differ across birth quarters?
groups = [athletes.loc[athletes["quarter_label"]==q, "age_at_games"].dropna()
          for q in QUARTER_LABELS]
f_stat, p_anova = stats.f_oneway(*groups)
print(f"One-way ANOVA (age ~ quarter):  F = {f_stat:.4f},  p = {p_anova:.4f}")
if p_anova >= 0.05:
    print("  -> No significant age difference (p>=0.05); age is NOT a confound.")
else:
    print("  -> Significant age difference found; age may be a covariate.")

print("\nAge summary by quarter:")
display(athletes.groupby("quarter_label")["age_at_games"]
        .agg(["mean","median","std"]).round(2))
```

The ANOVA rules out (or confirms) that the birth-quarter pattern is simply an artifact of age differences between birth cohorts. If F is not significant, the birth-timing interpretation is more credible. Specifically, the above ANOVA and violin plot indicate no significant age differences between birth quarters, suggesting that the birth-quarter pattern is unlikely to be attributed to athletes' age differences.

## Discussion and Conclusion

In our project, we investigated whether U.S. Winter Olympic athletes (2022 & 2026 Games) differ from the general American population in their birth month distribution. Our research question considered how these athletes' birth months compare to the U.S. birth baseline from 1995-2005. Additionally, we aimed to find whether the Relative Age Effect, the tendency for athletes born earlier in the year to be overrepresented, would appear at the elite Olympic level.

Contrary to our hypothesis, we found that Q1 (January-March) athletes were not overrepresented. Rather, surprisingly, both Q1 and Q4 athletes were slightly underrepresented relative to the U.S. birth baseline, while Q2 and Q3 athletes were overrepresented. Neither the quarterly chi-square test nor the one-tailed z-test showed statistical significance in regards to Q1 overrepresentation. This suggests that the classic early birth advantage that we researched was documented in youth sports may not necessarily persist to the Olympic level. These findings are consistent with the possibility that factors beyond birth months, such as training and experience, become increasingly important at the professional Olympic level.

We observed that January showed the largest negative deviation from its expected count. Mid-year months such as June tended to be overrepresented. Furthermore, the sport-level heatmap revealed that the birth quarter pattern was not uniform across disciplines. In other words, some sports showed the Relative Age Effect more than others, consistent with prior research where RAE effects were stronger in hockey and alpine skiing while weaker in figure skating.

One noteworthy finding is that birth quarter did not meaningfully predict medal success. While Q3 appeared to have a slightly higher medal count, the confidence intervals across all quarters overlapped a considerable amount. The one-way ANOVA also confirmed that age at the Games did not differ significantly across birth quarters, ruling out age by month as a confounding factor in Olympic participation.

The most notable limitations we faced were related to sample size. With only around 450 athlete-Games appearances spread across 12 months, the per-cell counts were small, making month-level chi-square tests noisy. Pooling into quarters helped, but the overall sample remains modest for detecting subtle effects. Additionally, our dataset captures only athletes who made the Olympic team, meaning we observe the survivors of a long selection process and cannot see how birth timing may have filtered athletes earlier in youth pipelines. Different sports also use different selection year cutoffs, which, when pooled together, likely attenuates any RAE signal present within individual sports.

Another limitation of our project is that our data is limited to 2 iterations of Winter Olympics. With only 2 events being included, it remains a possibility that an external variable could have affected which athletes were selected or were available to participate in each of the 2 Games. For instance, the 2022 Winter Olympics could potentially have been impacted by COVID-19 regulations and disruptions, which may have influenced athlete participation and selection for the Games.

In the future, we intend to expand the project to conduct analysis on U.S. Summer Olympics athletes, specifically 2020 and 2024. We aim to investigate whether the absence of the Relative Age Effect is unique to U.S. Winter Olympics, or whether this phenomenon also occurs in Summer Olympics.
